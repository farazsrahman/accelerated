```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000  # You can change N to the desired number of points

# Compute angles for N points evenly spaced on the top semicircle [0, pi]
theta = np.linspace(0, np.pi, N)
# Points on the unit circle
x = np.cos(theta)
y = np.sin(theta)

# Set up the plot
fig, ax = plt.subplots(figsize=(6, 4))
# Plot the semicircle
circle_theta = np.linspace(0, np.pi, 100)
ax.plot(np.cos(circle_theta), np.sin(circle_theta), 'k--', label="Unit semicircle")

# Plot points on semicircle
ax.scatter(x, y, color='blue', label='Points on semicircle')

# Project them down to the x-axis
ax.scatter(x, np.zeros_like(x), color='red', label='Projection on x-axis')

# Draw lines for projections
for xi, yi in zip(x, y):
    ax.plot([xi, xi], [yi, 0], 'gray', linestyle=':', linewidth=1)

ax.set_aspect('equal')
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-0.1, 1.1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.set_title(f"{N} points on the top semicircle and their projections to the x-axis")
plt.show()

# Plot a histogram of the x-axis projections (the red points)
plt.figure(figsize=(6, 4))
counts, bins, patches = plt.hist(x, bins=100, color='red', alpha=0.7, rwidth=0.85, edgecolor='black', density=True)

# Plot the ground truth arcsin distribution for projection onto x-axis
# Arcsin distribution PDF: f(x) = 1 / (pi * sqrt(1 - x^2)) for x in [-1, 1]
x_grid = np.linspace(-1, 1, 500)
arcsin_pdf = 1.0 / (np.pi * np.sqrt(1 - x_grid**2))
plt.plot(x_grid, arcsin_pdf, 'b-', label='Arcsin PDF (ground truth)', linewidth=2)

plt.xlabel('Projected x value')
plt.ylabel('Density')

plt.title('Histogram of Projected x-axis Values with Arcsin Distribution')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()

```
